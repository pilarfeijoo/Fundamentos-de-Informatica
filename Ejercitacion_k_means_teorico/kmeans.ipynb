```python
stock_data = pd.read_csv("/Users/pilarfeijoo/Desktop/fundamentos de informatica/Fundamentos-de-Informatica/tablas de datos/dataset_clustering_teorico.csv")
stock_data.head()
```

```python

stock_data= stock_data.set_index('Unnamed: 0')
stock_data.head()
```

```python
len(stock_data)
```

```python

stock_data.drop_duplicates(inplace=True)
stock_data.dropna(inplace=True)
```

```python

len(stock_data)
```

```python

#Normalizar datos
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
stock_data_normalizado = scaler.fit_transform(stock_data)
```

```python
from sklearn.cluster import KMeans

#Agrupar datos
k = 14 #definimos la cantidad de clusters
kmeans = KMeans(n_clusters = k, init="random", n_init=10, max_iter=300, random_state=123457) #tomamos los centroides de forma aleatoria y definimos un máximo de 300 iteraciones
kmeans.fit(stock_data_normalizado)  #aplicamos el método a nuestros datos
```

```python

print(kmeans.labels_)
print(kmeans.cluster_centers_ )
```

```python
from sklearn.cluster import KMeans
import seaborn as sns

def probar_k(df, max):
    k = 0
    datos = {"K": [], "Inercia": []}
    valores_k = []
    valores_incercia = []

    while k < max:
        k += 1
        kmeans = KMeans(n_clusters = k, init="random", n_init=10, max_iter=300, random_state=123457)
        kmeans.fit(df)
        inercia = kmeans.inertia_ 
        valores_k.append(k)
        valores_incercia.append(inercia)
        
    datos["K"] = valores_k
    datos["Inercia"] = valores_incercia

    k_vs_inertia = pd.DataFrame(datos)
    print(k_vs_inertia)
    
    inertia_vs_k = sns.pointplot(data = k_vs_inertia, x="K", y="Inercia")
    inertia_vs_k

probar_k(stock_data_normalizado, 20)
```
